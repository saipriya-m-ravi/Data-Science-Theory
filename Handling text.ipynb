```python
import pandas as pd
housing=pd.read_csv("housing.csv")
```

```python
housing_cat=housing[["ocean_proximity"]]
housing_cat.head()
```

Encoding => Categorical to numerical

```python
## 1.Scikit-Learn OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder=OrdinalEncoder()
housing_cat_encoded=ordinal_encoder.fit_transform(housing_cat)
housing_cat_encoded
```

```python
ordinal_encoder.categories_
```

Here ML algo assumes two nearby values are more similar than two distant values.For ordered categories its fine.

Solution - one hot encoding


```python
## 2.Scikit-Learn OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
cat_encoder=OneHotEncoder()
housing_cat_1hot=cat_encoder.fit_transform(housing_cat)   # => Output is a SciPy sparse matrix
housing_cat_1hot
```

```python
housing_cat_1hot.toarray()
```

```python
cat_encoder.categories_
## OneHotEncoder remembers the categories its trained on. It will detect unknown category and raise an exception
```

If there are a lot of categories for example "Profession", then one hot encoding will result in a huge matrix which will degrade the performance. Then better replace categorical input with useful numerical features.

eg: ocean_proximity to "distance to the ocean"

country code to country's population and gdp per capita
