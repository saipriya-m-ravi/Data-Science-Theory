```python
## 1. Get rid of those instances (rows)
housing.dropna(subset=["total_bedrooms"],inplace=True)
```

```python
## 2. Get rid of the attribute (column)
housing.drop("total_bedrooms",axis=1)
```

```python
## 3. Imputation - Set missing values to zero,mean,median etc
median=housing["total_bedrooms"].median()
housing["total_bedrooms"].fillna(median,inplace=True)
```

```python
## Scikit-Learn class:SimpleImputer (It stores median of each feature so that we can impute missing values on training,validation and new data set too)
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy="median")
housing_num=housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)

## Safer to apply imputer to all numerical attrs because missing values might come to other attrs too once system goes live.
imputer.statistics_    => median of all attrs is stored in this var

X=imputer.transform(housing_num)
```
