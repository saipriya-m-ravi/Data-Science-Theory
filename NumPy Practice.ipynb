```python
# How to make a python function that handles scalars to work on numpy arrays?
# Convert the function maxx that works on two scalars, to work on two arrays.

def maxx(x, y):
    """Get the maximum of two items"""
    if x >= y:
        return x
    else:
        return y

maxx(1, 5)
#> 5

pair_max=np.vectorize(maxx,otypes=[float])


a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
pair_max(a, b)
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

print(softmax(sepallength))
```

```python
def softmax(x):
    e_x = np.exp(x)
    return e_x / e_x.sum(axis=0)

print(softmax(sepallength))
```

The softmax function is a function that turns a vector of K real values into a vector of K real values that sum to 1. The input values can be positive, negative, zero, or greater than one, but the softmax transforms them into values between 0 and 1, so that they can be interpreted as probabilities.

e^Zi/(sigma j=1 to k e^Zj)

Used in neural networks in multiclass classification.

```python
def counts_of_all_values_rowwise(arr2d):
    # Unique values and its counts row wise
    num_counts_array = [np.unique(row, return_counts=True) for row in arr2d]

    # Counts of all values row wise
    return([[int(b[a==i]) if i in a else 0 for i in np.unique(arr2d)] for a, b in num_counts_array])

counts_of_all_values_rowwise(arr)
```

```python
# How to generate one-hot encodings for an array in numpy?
np.random.seed(101)
arr = np.random.randint(1,4, size=6)
len(arr)
```

```python
encoding=[[0,0,0] for _ in range(len(arr))]
for i,x in enumerate(arr):
  encoding[i][x-1]=1

np.array(encoding).astype('float')
```

```python
def one_hot_encoding(arr):
  uniques=np.unique(arr)
  encoding=np.zeros((arr.shape[0],uniques.shape[0]))
  for i,k in enumerate(arr):
    encoding[i,k-1]=1
  return encoding

one_hot_encoding(arr)
```

```python
# How to fill in missing dates in an irregular series of numpy dates?
dates=np.arange(np.datetime64('2018-02-01'),np.datetime64('2018-02-25'),2)
dates
```

```python
out=[]
for date,d in zip(dates,np.diff(dates)):
  out.append(np.arange(date,date+d))

filled_dates=np.array(out).reshape(-1)
final=np.hstack([filled_dates,dates[-1]])
final
```

```python
# How to create strides from a given 1D array?
# From the given 1d array arr, generate a 2d matrix using strides, with a window length of 4 and strides of 2, like [[0,1,2,3], [2,3,4,5], [4,5,6,7]..]

arr=np.arange(15)
arr
```

```python
out=[]
window_len=4
stride=2
for i in range(0,len(arr)-window_len,stride):
  out.append(arr[i:i+window_len])
np.array(out)
```
